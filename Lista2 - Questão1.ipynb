```python
train_x = train_clean.iloc[0:: , 1:15].copy()
train_y = train_clean.iloc[0:: , 15:16].copy()

test_x = test_clean.iloc[0:: , 1:15].copy()
test_y = test_clean.iloc[0:: , 15:16].copy()
```

## Vizinhos x Acurácia

```python
n = []
accur = []

for i in range(2,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_x.values,train_y.values.ravel())
    YtestPred = knn.predict(test_x)
    
    n.append(i)
    accur.append(accuracy_score(YtestPred,test_y.values.ravel()))
    
```

```python
fig = plt.plot(n, accur, color='red', label="test", marker='o')

plt.xlabel('Nº Vizinhos')
plt.ylabel('Acurácia')
plt.title("Acurácia x Nº de Vizinhos")
plt.show()

fig.savefig('acurXneighbor.png')

fig = plt.figure()
```

```python


fig, ax = plt.subplots( nrows=1, ncols=1 )  # create figure & 1 axis
ax.plot(n, accur, color='red', label="test", marker='o')
ax.title.set_text('Acurácia x Nº de Vizinhos')
ax.set_xlabel('Nº de Vizinhos')
ax.set_ylabel('Acurácia')
fig.savefig('figure2.png')   # save the figure to file
fig.show()
```

## Distância

```python
n = []
accur = []

for i in range(2,30):
    knn = KNeighborsClassifier(n_neighbors=i, weights="distance")
    knn.fit(train_x,train_y)
    YtestPred = knn.predict(test_x)
    
    n.append(i)
    accur.append(accuracy_score(YtestPred,test_y))
    
```

```python

fig, ax = plt.subplots( nrows=1, ncols=1 )  # create figure & 1 axis
ax.plot(n, accur, color='red', label="test", marker='o')
ax.title.set_text('Acurácia x Nº de Vizinhos (weights)')
ax.set_xlabel('Nº de Vizinhos')
ax.set_ylabel('Acurácia')
fig.savefig('figure3.png')   # save the figure to file
fig.show()
```

## Acurácia vs Número de Instâncias

### A partir da base de dados sera gerado fatias que correspondem a 20%, 40%, 60%, 80% E 100% do tamanho original, o número de vizinhos sera fixado em 15 com pesos baseados na distancia

```python
adult_full = pd.read_csv('adult-full-clean.csv', sep=",")  
adult_full_20 = adult_full.sample(frac=0.2, random_state=100)
adult_full_40 = adult_full.sample(frac=0.4, random_state=100)
adult_full_60 = adult_full.sample(frac=0.6, random_state=100)
adult_full_80 = adult_full.sample(frac=0.8, random_state=100)

adult_all = [adult_full_20, adult_full_40, adult_full_60, adult_full_80, adult_full]
for i in adult_all:
    print(i["age"].size)
```

### Cada sub conjunto da base (20, 40, 60, 80 e 100%) será dividido entre 70% para treinamento e 30% para testes

```python
tamanhos = [20, 40, 60, 80, 100]
trains = []
tests = []

for i in adult_all:

    train = i.sample(frac=0.7, random_state=100) #random state is a seed value
    test  = i.drop(train.index)

    trains.append(train)
    tests.append(test)


for i in range(0, len(trains)):
    print("Subconjuno {} %, divisão 70-30: {}-{}  ".format(tamanhos[i], trains[i]["age"].size,tests[i]["age"].size ))
```

```python
train_x = trains[0].iloc[0:: , 0:15].copy()
train_y = trains[0].iloc[0:: , 14:15].copy()


test_x = tests[0].iloc[0:: , 0:15].copy()
test_y = tests[0].iloc[0:: , 14:15].copy()
```

### Fixaremos o número de vizinhos em 15, considerando distancias uniformes

```python

accur = []
accur_weights = []

for i in range(0, len(trains)):
    
    train_x = trains[i].iloc[0:: , 0:15].copy()
    train_y = trains[i].iloc[0:: , 14:15].copy()


    test_x = tests[i].iloc[0:: , 0:15].copy()
    test_y = tests[i].iloc[0:: , 14:15].copy()

        
    
    knn = KNeighborsClassifier(n_neighbors=20)
    knn.fit(train_x.values,train_y.values.ravel())
    YtestPred = knn.predict(test_x)
    
    accur.append(accuracy_score(YtestPred,test_y.values.ravel()))

```

```python
fig, ax = plt.subplots( nrows=1, ncols=1 )  # create figure & 1 axis
ax.plot(tamanhos, accur, color='red', label="test", marker='o')
ax.title.set_text('Acurácia x Nº de Instâncias')
ax.set_xlabel('Nº de Instâncias')
ax.set_ylabel('Acurácia')
fig.savefig('figure4.png')   # save the figure to file
fig.show()
```

### Fixaremos o número de vizinhos em 15, considerando pesos de acordo com a distância

```python
accur_weights = []

for i in range(0, len(trains)):
    
    train_x = trains[i].iloc[0:: , 0:15].copy()
    train_y = trains[i].iloc[0:: , 14:15].copy()


    test_x = tests[i].iloc[0:: , 0:15].copy()
    test_y = tests[i].iloc[0:: , 14:15].copy()

        
    
    knn = KNeighborsClassifier(n_neighbors=20, weights="distance")
    knn.fit(train_x.values,train_y.values.ravel())
    YtestPred = knn.predict(test_x)
    
    accur_weights.append(accuracy_score(YtestPred,test_y.values.ravel()))
```
